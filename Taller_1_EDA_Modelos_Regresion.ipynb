```python
import pandas as pd
pd.set_option('display.max_columns', None)  # Show all columns

df = pd.read_csv('listings.csv.gz', compression='gzip')
df.head()
```

```python
df.shape
```

# Secci√≥n 1: Inspecci√≥n de datos

```python
# --- Paso 1: Inspecci√≥n inicial del dataset ---

# 1. Dimensiones del dataset (filas x columnas)
print("Dimensiones del dataset:", df.shape)

# 2. Informaci√≥n general (tipos de datos y valores no nulos)
print("\n--- Info general ---")
df.info()

# 3. Porcentaje de valores nulos por columna
print("\n--- Porcentaje de valores nulos por columna ---")
nulls = df.isnull().mean().sort_values(ascending=False) * 100
print(nulls.head(20))  # mostramos solo las 20 columnas con m√°s nulos

# 4. Verificar duplicados en 'id'
print("\nDuplicados en ID:", df['id'].duplicated().sum())
```

# Secci√≥n 2: Conversi√≥n de tipo de datos

```python
# --- Paso 2: Conversi√≥n de tipos de datos ---

# 1. Convertir fechas
date_cols = ['last_scraped', 'host_since', 'first_review', 'last_review']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Limpiar columna 'price'
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# 3. Limpiar porcentajes (host_response_rate y host_acceptance_rate)
percent_cols = ['host_response_rate', 'host_acceptance_rate']
for col in percent_cols:
    if col in df.columns:
        df[col] = df[col].str.replace('%', '', regex=False).astype(float)

# Revisar que qued√≥ bien
print(df[['price', 'host_response_rate', 'host_acceptance_rate']].head())
print(df[date_cols].dtypes)
```

# Secci√≥n 3: An√°lisis de valores nulos

```python
# --- Paso 3: An√°lisis de valores nulos ---

nulls = df.isnull().mean().sort_values(ascending=False) * 100
print("Porcentaje de valores nulos por columna:\n")
print(nulls)

# Visualizaci√≥n r√°pida de las columnas con m√°s del 30% de nulos
import matplotlib.pyplot as plt

nulls[nulls > 30].plot(kind='bar', figsize=(12,5), title="Columnas con m√°s del 30% de nulos")
plt.show()
```

# Secci√≥n 4: Eliminar columnas nulas NO relevantes

```python
# --- Paso 4: Eliminaci√≥n de columnas irrelevantes o con demasiados nulos ---

cols_to_drop = [
    "calendar_updated",
    "host_neighbourhood",
    "neighborhood_overview",
    "neighbourhood",
    "host_about",
    "license"
]

df = df.drop(columns=cols_to_drop)

print(f"Dataset despu√©s de eliminar columnas: {df.shape}")
```

# Secci√≥n 5:

```python
# --- Paso 5: Imputaci√≥n de valores nulos en columnas num√©ricas ---

# Separamos num√©ricas y categ√≥ricas
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Revisamos cu√°ntos nulos tienen las num√©ricas
nulls_num = df[num_cols].isnull().mean() * 100
print("Nulos en num√©ricas (%):")
print(nulls_num[nulls_num > 0].sort_values(ascending=False))

# Imputamos con mediana (m√°s robusta ante outliers)
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

print("\nDespu√©s de imputar, nulos en num√©ricas:", df[num_cols].isnull().sum().sum())
```
